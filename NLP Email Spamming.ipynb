Cleaning Data 

```python
from nltk.corpus import stopwords
from nltk.stem import  PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
lemmatizer = WordNetLemmatizer()
stopw = set(stopwords.words("english"))
stopw.update(".", "?", "(", ")", ",", "-", "'", '"',"/","2","1","3","4","5","6","7","8","9","10","&")
```

```python
def text_to_words(raw_text):
    
    pos_t = (pos_tag(word_tokenize(raw_text)))
    words = raw_text.lower()
    
    lemmatizer.lemmatize(words)
    words = words.split()
    words = [word for word in words if not word.startswith('ham')]
    words = [word for word in words if not word.startswith('spam')]
    
    meaningful_words = [w for w in words if not w in stopw]
    
    return (" ".join(meaningful_words))
```

```python
temp_df['clean_text'] = temp_df['text'].apply(text_to_words)
temp_df
```
